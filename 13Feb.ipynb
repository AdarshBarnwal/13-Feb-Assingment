```python
Q4.
Ans:-

Because,Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence
of a list/dictionary is invalid or does not exists.

The two types of exceptions raised are:
    
KeyError
IndexError

#KeyError
 If a key we are trying to access is not found in the dictionary, a key error exception is raised.
    
for example:-

import logging

logging.BasicConfig(filenmae="output.log",level=logging.INFO)

try:
    dic={1:"Rahul",2:"Rohan"}
    logging.error("i am trying to print data %s".format(dic[4]))
    
except LookupError as e:
    
    logging.error("key not found {}".format((e)))
    

#indexError

When we are trying to access an index (sequence) of a list that does not exist in that list or is out of range
of that list,an index error is raised.

for example:-

import logging

logging.basicConfig(filename="output.log"level=logging.INFO)

try:
    a = ['a','b','c']
    
    logging.info("i am trying to print a[4]")
    
    print (a[4])
    
except LookupError:
    logging.error("index Error Exception Raised,list index out of range")
```
